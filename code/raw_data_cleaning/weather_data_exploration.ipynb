```python
import pandas as pd
```

```python
data = pd.read_csv("../../data/raw_data/weather/weather.csv")
```

```python
print(data.shape)
```

```python
print(data.isna().any().any())
```

```python
data["TemperatureAboveGroundCelcius"] = data["TemperatureAboveGround"] - 273.15
```

```python
netherlands_data = data[data["iso3166-2"].str.startswith("NL")]
```

```python
data_nl_statistics = netherlands_data.groupby("iso3166-2").agg(["min", "mean", "median", "max"])
data_nl_statistics["TemperatureAboveGroundCelcius"]["mean"].plot.bar().legend(loc = "lower left")
```
